```python
import copy
import numpy as np

VISITED_POSITION = 'X'

def drawBoard(board):
    return "\n".join(["".join(line) for line in board])

def findGuardCurrentPositionDirection(board):
    for i, row in enumerate(board):
        for j, column in enumerate(row):
            match column:
                case '^':
                    return (i,j), 'up'
                case '>':
                    return (i,j), 'right'
                case '<':
                    return (i,j), 'left'
                case 'v':
                    return (i,j), 'down'

TURNS = {
        'up':'right',
        'right':'down',
        'down': 'left',
        'left': 'up'
    }

def turn(currentDirection):
    return TURNS[currentDirection]

MOVES = {
        'up': (-1,0),
        'right':(0,1),
        'down': (1,0),
        'left': (0,-1)
    }

def moveForward(currentPosition, direction):
    move = MOVES[direction]
    return (
        currentPosition[0]+move[0],
        currentPosition[1]+move[1],
    )
    
def checkNextStep(board, position, direction):
    '''
    Check if the next step is available, return the position and the direction
    if theres an obstacle, stay in position and turn, if no obstacle, move one step
    return False if left the board. 
    
    '''
    boardHeight = len(board[0])
    boardWidth = len(board)
    
    currentDirection = direction
    #if were at an edge facing out, return done
    if (
        ((position[0] == 0) and (currentDirection == 'up')) or #top
        ((position[0] == boardHeight-1) and (currentDirection == 'down')) or #bottom
        ((position[1] == 0) and (currentDirection == 'left')) or #left
        ((position[1] == boardWidth-1) and (currentDirection == 'right')) #right
    ):
        return (position), False
    
    # get next position, if moving forward
    nextPosition = moveForward(position, currentDirection)
    if board[nextPosition[0]][nextPosition[1]] == '#':
        return position, turn(currentDirection)
    else:
        return nextPosition, currentDirection

GUARDS = {
        'up': '^',
        'right': '>',
        'down': 'v',
        'left': '<'
    }

def moveGuard(board, previousPosition, currentPosition = None, currentDirection = None):
     currentBoard = copy.deepcopy(board)
     currentBoard[previousPosition[0]][previousPosition[1]] = VISITED_POSITION
     if currentDirection:
        currentBoard[currentPosition[0]][currentPosition[1]] = GUARDS[currentDirection]
     return currentBoard
    
def countPositionsVisited(board):
    concatBoard = sum(board, [])
    return concatBoard.count(VISITED_POSITION)

def trackGuard(board, trackLoop = True):
    positionAndDirectionVisited = []
    currentBoard = copy.deepcopy(board)
    currentPosition, currentDirection = findGuardCurrentPositionDirection(currentBoard)
    positionAndDirectionVisited.append((currentPosition, currentDirection))
    while currentDirection:
        previousPosition = currentPosition
        currentPosition, currentDirection = checkNextStep(currentBoard, currentPosition, currentDirection)
        if ((currentPosition, currentDirection) in positionAndDirectionVisited):
            return 'LOOP'
        positionAndDirectionVisited.append((currentPosition, currentDirection))
        currentBoard = moveGuard(currentBoard, previousPosition, currentPosition, currentDirection)
    if trackLoop:
        return 'NOT A LOOP'
    previousPosition = currentPosition
    currentBoard = moveGuard(currentBoard, previousPosition)
    
    print("\n\n")
    print(drawBoard(currentBoard))
    print("\n\n")
    # print(f'Guard visited {countPositionsVisited(currentBoard)} positions!')
    return positionAndDirectionVisited

from tqdm.notebook import tqdm

def addObstacleForLoop(board, path=None):
    loopyObstacles = []
    originalBoard = copy.deepcopy(board)
    print(len(board))
    for i, row in tqdm(enumerate(originalBoard)):
        print(f'*** row {i} ***')
        for j, column in enumerate(row):
            boardCopy = copy.deepcopy(originalBoard)
            if column == '#' or column == '^':
                continue
            elif path and ((i,j) in [element[0] for element in path]): 
                boardCopy[i][j] = '#'
                # print(drawBoard(boardCopy))
                result = trackGuard(boardCopy, trackLoop=True)
                if result == 'LOOP':
                    loopyObstacles.append((i,j))
            else:
                continue
    return loopyObstacles
                    
        
    
```

```python
with open('day6_input.txt') as file:
    board = [row.strip() for row in file.readlines()]

# board = [
# '....#.....',
# '....^....#',
# '..........',
# '..#.......',
# '.......#..',
# '..........',
# '.#........',
# '........#.',
# '#.........',
# '......#...',
# ]

board = [list(line) for line in board]
board = np.array(board)
path = trackGuard(board, trackLoop=False)
loopyObstacles = addObstacleForLoop(board, path = path)

print(f'There are {len(loopyObstacles)} possible obstacles to cause a loop')
```
