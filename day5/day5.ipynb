```python
with open('day5_input.txt') as file:
    lines = [line.strip() for line in file.readlines()]

# lines = [
# '47|53',
# '97|13',
# '97|61',
# '97|47',
# '75|29',
# '61|13',
# '75|53',
# '29|13',
# '97|29',
# '53|29',
# '61|53',
# '97|53',
# '61|29',
# '47|13',
# '75|47',
# '97|75',
# '47|61',
# '75|61',
# '47|29',
# '75|13',
# '53|13',
# '',
# '75,47,61,53,29,1',
# '97,61,53,29,13',
# '75,29,13',
# '75,97,47,61,53',
# '61,13,29',
# '97,13,75,29,47',
# ]

# split input
rules = []
updates = [] 
rulesQ = True
for line in lines:
    if line == "":
        rulesQ = False
    elif rulesQ:
        rules.append(line)
    else:
        updates.append(line)

# process rules     
rules_dict = {}
for line in rules:
    before, after = line.split('|', 1)
    if int(before) in rules_dict:
        rules_dict[int(before)].append(int(after))
    else: 
        rules_dict[int(before)] = [int(after)]
        
def validUpdateQ(update, rules_dict=rules_dict):
    for i, number in enumerate(update):
        if number in rules_dict:
            upstream = set(update[:i])
            # check if any numbers are upstream that shouldnt
            if bool(upstream & set(rules_dict[number])):
                return False, number
    return True, 0
                
def getMiddleElement(update):
    if len(update)%2==0:
        print(f'{update} has an even number of elements!')
    return update[(len(update)-1)//2]

def getMiddles(updates):
    middles = []
    for update in updates:
        splitUpdate = [int(num) for num in update.split(',')]
        print(splitUpdate)
        validQ, invalidPageNumber = validUpdateQ(splitUpdate)
        if validQ:
            print('Valid')
            middles.append(getMiddleElement(splitUpdate))
        else:
            print(f'inValid ({invalidPageNumber})')
    return middles

def findFirstValidPositionForPage(splitUpdate, page, rules_dict=rules_dict):
    pagesToPreced = rules_dict[page]
    for i in range(len(splitUpdate)):
        if splitUpdate[i] in pagesToPreced:
            return i
    return False

def fixInvalid(splitUpdate, faultyPage, rules_dict=rules_dict):
    currentUpdate = splitUpdate.copy()
    currentUpdate.remove(faultyPage)
    positionToMoveTo = findFirstValidPositionForPage(currentUpdate, faultyPage, rules_dict)
    currentUpdate.insert(positionToMoveTo, faultyPage)
    return currentUpdate
        

def iterativelyFixInvalid(splitUpdate, rules_dict=rules_dict):
    currentUpdate = splitUpdate.copy()
    validQ, invalidNumber = validUpdateQ(currentUpdate)
    counter = 0
    while (not validQ) and (counter < len(splitUpdate)):
        currentUpdate = fixInvalid(currentUpdate, invalidNumber, rules_dict)
        validQ, invalidNumber = validUpdateQ(currentUpdate)
    if validQ:
        return currentUpdate, True
    else:
        return splitUpdate, False

def getInvalidsMiddles(updates):
    middles = []
    for update in updates:
        splitUpdate = [int(num) for num in update.split(',')]
        print(splitUpdate)
        validQ, invalidPageNumber = validUpdateQ(splitUpdate)
        if validQ:
            print('Valid')
        else:
            print(f'inValid ({invalidPageNumber})')
            currentUpdate, validNowQ = iterativelyFixInvalid(splitUpdate)
            if validNowQ:
                middles.append(getMiddleElement(currentUpdate))
            else:
                print(f'Cant Correct!')
    return middles        

            
```

```python
middles = getMiddles(updates)
print(f'the sum of middles of valid lists is: {sum(middles)}')
```

```python
middles = getInvalidsMiddles(updates)
print(f'the sum of middles of valid lists is: {sum(middles)}')    
```
